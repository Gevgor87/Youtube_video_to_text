```python
import whisper
import torch 
import os
```

```python

device = "cuda" if torch.cuda.is_available() else "cpu"

whisper_model = whisper.load_model("large", device=device) #or medium model
```

```python
from pytube import YouTube
```

```python
def video_to_audio(video_URL, destination, final_filename):

  # Get the video
  video = YouTube(video_URL)

  # Convert video to Audio
  audio = video.streams.filter(only_audio=True).first()

  # Save to destination
  output = audio.download(output_path = destination)

  _, ext = os.path.splitext(output)
  new_file = final_filename + '.mp3'

  # Change the name of the file
  os.rename(output, new_file)

# Video to audio
video_URL = 'https://www.youtube.com/watch?v=Hi0zbU3DW6o'
destination = "."
final_filename = "73_Questions_With_Kylie_Jenner"
video_to_audio(video_URL, destination, final_filename)
```

```python
audio_file = "73_Questions_With_Kylie_Jenner.mp3"
result = whisper_model.transcribe(audio_file)
```

```python
print(result["text"])
```

```python
with open(f"{final_filename}.txt", "w") as file:
    file.write(result["text"])
```
